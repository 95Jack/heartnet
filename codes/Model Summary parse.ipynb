```python
import pandas as pd

summary = "/media/taufiq/Data1/heart_sound/heartnet/model"
```

```python
with open(summary,'r') as fID:
    x = fID.read()
x = x.replace('[0]','')
x = x.split('\n')[0::2]
x = [[each for each in row.split(' ') if not each=='']  for row in x]
del x[0][-1]
x[1].append('')
x[-1].append('')
x[-2].append('')
```

```python
for each in x:
    print(len(each))
```

```python
pd.DataFrame(x).to_csv('summary.csv')
```
